# [TR_paciente_Delete]

```sql
/****** Object:  Trigger [dbo].[TR_paciente_Delete]    Script Date: 19/10/2023 7:23:36 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE TRIGGER [dbo].[TR_paciente_Delete]
ON [dbo].[paciente]
AFTER DELETE
AS
BEGIN
    INSERT INTO movimientos(tipo_movimiento, fecha, hora, dato_anterior)
    SELECT 'Eliminación', GETDATE(), GETDATE(), d.nombre
    FROM deleted d;
END;
GO
ALTER TABLE [dbo].[paciente] ENABLE TRIGGER [TR_paciente_Delete]
GO
```

# [TR_paciente_Insert]

```sql
/****** Object:  Trigger [dbo].[TR_paciente_Insert]    Script Date: 19/10/2023 7:23:36 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE TRIGGER [dbo].[TR_paciente_Insert]
ON [dbo].[paciente]
AFTER INSERT
AS
BEGIN
	INSERT INTO movimientos (tipo_movimiento,fecha,hora)
	SELECT 'insercion',GETDATE(),GETDATE()
	FROM inserted;
END;
GO
ALTER TABLE [dbo].[paciente] ENABLE TRIGGER [TR_paciente_Insert]
GO
```

# [TR_paciente_Update]

```sql
/****** Object:  Trigger [dbo].[TR_paciente_Update]    Script Date: 19/10/2023 7:23:36 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE TRIGGER [dbo].[TR_paciente_Update]
ON [dbo].[paciente]
AFTER UPDATE
AS
BEGIN
    INSERT INTO movimientos (tipo_movimiento, hora, fecha, dato_nuevo, dato_anterior)
    SELECT 'actualizacion', GETDATE(), GETDATE(),d.nombre, i.nombre
    FROM deleted d
    INNER JOIN inserted i ON d.Id = i.Id;
END;
GO
ALTER TABLE [dbo].[paciente] ENABLE TRIGGER [TR_paciente_Update]
GO
```
